# 预处理

```python
import numpy as np
import pandas as pd


frame = pd.read_csv('data.csv',header=0,index_col=0)
# 处理错误的学校名称
schools = frame['school']
schools[schools=='& Engineering; Northeastern\nUniversity'] = \
'School of Computer Science & Engineering; Northeastern University'
schools[schools=='& Simulation; Northeastern\nUniversity'] = \
'Research Center of Control & Simulation; Northeastern University'
schools = schools.str.replace('\n',' ')
schools[schools.str.contains(';')] = 'NORTHEASTERN UNIVERSITY'# 这个大学把学院都写上了，去掉它的学院
# 将学校名称变为大写
schools = schools.str.upper()
# 去掉空格
schools = schools.str.replace(' ','')
# 去掉学院
schools = schools.str.replace('SCHOOLOF.*,','')
frame['school'] = schools

# 处理奖项
# O奖:0,F奖:1,M奖:2,H奖:3,S奖:4,U奖:5
awards = frame['award']
awards[awards.str.contains('Outstanding')] = 'O'
awards[awards == 'Finalist'] = 'F'
awards[awards == 'Meritorious Winner'] = 'M'
awards[awards == 'Honorable Mention'] = 'H'
awards[awards == 'Successful Participant'] = 'S'
awards[awards == 'Unsuccessful'] = 'U'
awards = awards.astype('str')

# 有的队伍名字为空，改为''(空字符串)
frame['name'][frame['name'].isnull()] = ''
```

## 各学校参赛队数和获奖情况（仅展示到西工大）

```python
# 生成数据透视表
school_awards = frame.pivot_table(values='name',index='school',columns='award',aggfunc=len,fill_value=0,margins=True)
# 调整列顺序
school_awards = pd.DataFrame(school_awards,columns=['O','F','M','H','S','U','All'],dtype='int32')
school_awards = school_awards.sort_values(by=['All','O','F','M','H','S','U'],ascending=False)
school_awards.index = [school_map.get(i,i) for i in school_awards.index]
morethan80 = school_awards[school_awards['All']>80]
school_awards[:'西工大']
```

# 统计获奖率（参赛队大于30以上的学校按照M以上比例排名）

```python
table = school_awards.sort_values(['O','All','O','F','M','H','S','U'],ascending=False)
rate = pd.DataFrame(table,columns = ['O','F','M','H','S','U','All'])
for c in ['O','F','M','H','S','U']:
    rate[c] = rate[c]/rate['All']*100

rate['O以上'] = rate['O']
rate['F以上'] = rate['O']+rate['F']
rate['M以上'] = rate['M']+rate['O']+rate['F']
rate['H以上'] = rate['M']+rate['O']+rate['F']+rate['H']
rate = rate.sort_values('M以上',ascending=False)
rate = rate[rate['All']>=30].sort_values('M以上',ascending=False)
rate.ix[:,'All':]
```

```python
total = rate.ix['总计']
total = total[:'U']

plt.pie(total,labels=['O','F','M','H','S','U'],
        explode=[0.3, 0.1,0.1,0,0,0],
        autopct='%2.2f',
        )
plt.title('Winning Percentage')
plt.show()
```
